```python
df.drop_duplicates()
```

```python
# replace � by blank space
df['ProductName']= df['ProductName'].str.replace('�','')
df['ProductName']
```

```python
df['CompanyName'].str.contains('�').sum()
```

```python
df['CompanyName']= df['CompanyName'].str.replace('�','')
```

```python
df['BrandName']= df['BrandName'].str.replace('�','')
```

```python
df.head()
```

```python
# convert dtype of CSFid into int64
df['CSFId']= df['CSFId'].fillna(0)
```

```python
df['CSFId']= df['CSFId'].astype('int32')
```

```python
# convert missing data into not available in CSF, brandname
df['CSF'] = df['CSF'].fillna('Data Not Available')
df['BrandName'] = df['BrandName'].fillna('Data Not Available')
```

```python
# convert dtype of dates columns into date format
df.info()
```

```python
df['MostRecentDateReported']= pd.to_datetime(df['InitialDateReported'])
```

```python
df
```

```python
df['InitialDateReported']= pd.to_datetime(df['MostRecentDateReported'])
```

```python
df['ChemicalUpdatedAt']= pd.to_datetime(df['ChemicalUpdatedAt'])
```

```python
df['ChemicalCreatedAt']= pd.to_datetime(df['ChemicalCreatedAt'])
```

```python
# drop column ChemicalDateRemoved
df= df.drop('ChemicalDateRemoved', axis=1)
```

```python
df
```

```python
# add column which contains active or inactive based on DiscontinuedDate
## Feature Engineering

df['ProductStatus'] = np.where(df['DiscontinuedDate'].notna(), 'inactive', 'active')
```
